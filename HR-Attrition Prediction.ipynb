```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

# ------------------------------- Decision Tree Classifier--------------------------------

```python
from sklearn.tree import DecisionTreeClassifier

dct1 = DecisionTreeClassifier(max_depth=8,min_samples_split=100,min_samples_leaf=10)
dct1.fit(x_train, y_train)


print(f"Training Accuracy: {dct1.score(x_train, y_train)}")
print(f"Test Accuracy: {dct1.score(x_test, y_test)}")
```

### employing Gridsearch to check for best parameters

```python
from sklearn.model_selection import GridSearchCV

dct=DecisionTreeClassifier()

parameters={"max_depth":range(4,50,3),
            "min_samples_split":range(20,150,5),
            "min_samples_leaf":range(10,50,5)}

gridSearch=GridSearchCV(dct,parameters,cv=5,scoring="accuracy")

gridSearch.fit(x,y)

print(f"Best Paramters: {gridSearch.best_params_}")

bestmodel=gridSearch.best_estimator_
```

```python


print(f"Training Accuracy: {bestmodel.score(x_train, y_train)}")
print(f"Test Accuracy: {bestmodel.score(x_test, y_test)}")
```

```python
parameters2={"max_depth":range(2,20,1),
            "min_samples_split":range(80,120,5),
            "min_samples_leaf":range(25,50,5)}

gridSearch2=GridSearchCV(dct,parameters2,cv=5,scoring="accuracy")

gridSearch2.fit(x,y)

print(f"Best Paramters: {gridSearch2.best_params_}")

bestmodel2=gridSearch2.best_estimator_


print(f"Training Accuracy: {bestmodel2.score(x_train, y_train)}")
print(f"Test Accuracy: {bestmodel2.score(x_test, y_test)}")
```

```python
dct2 = DecisionTreeClassifier(max_depth=4,min_samples_split=95,min_samples_leaf=30)
dct2.fit(x_train, y_train)


print(f"Training Accuracy: {dct2.score(x_train, y_train)}")
print(f"Test Accuracy: {dct2.score(x_test, y_test)}")
```

-- testing accuracy getting higher than training accuracy,

which  means the model is overfitting,

and it is happening due to small training dataset

### Pruning the model 

```python
dct3 = DecisionTreeClassifier(max_depth=4,min_samples_split=95,min_samples_leaf=30,ccp_alpha=0.01)
dct3.fit(x_train, y_train)


print(f"Training Accuracy: {dct3.score(x_train, y_train)}")
print(f"Test Accuracy: {dct3.score(x_test, y_test)}")
```

--its seems good to have 83.7% accuracy with 1% of gap between training and  testing accuracy


```python
Final_model=DecisionTreeClassifier(max_depth=4,min_samples_split=95,min_samples_leaf=30,ccp_alpha=0.01)
Final_model.fit(x_train, y_train)
```

### Evaluating the Final model with cross validation

```python
from sklearn.model_selection import cross_val_score
cv_scores = cross_val_score(Final_model, x, y, cv=10)
mean_cv_score = np.mean(cv_scores)
print("Mean CV Score with Best Model:", mean_cv_score)
```

---Mean Cross validation score is consistent with the model accuracy

### Confusion matrix

```python
from sklearn.metrics import confusion_matrix

y_pred = Final_model.predict(x_test)

confusion = confusion_matrix(y_test, y_pred)

print("Confusion Matrix:")
print(confusion)
```

```python
rfc1.set_params(warm_start=True, 
                  oob_score=True)

error_rate = {}

for i in range(10,1000,1):
    rfc1.set_params(n_estimators=i)
    rfc1.fit(x_train, y_train)
    oob_error = 1 - rfc1.oob_score_
    error_rate[i] = oob_error
```

```python
data3.head()
```

```python
x=data3.drop("Attrition",axis=1)
y=data3.Attrition
```

```python
## Scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x)
X = scaler.transform(x)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
from sklearn.svm import SVC
from sklearn import metrics

```

### Optimize Parameter C

```python
from sklearn.model_selection import cross_val_score

C_range=list(range(1,26))
acc_score=[]
for c in C_range:
    svc = SVC(kernel='linear', C=c)
    scores = cross_val_score(svc, X, y, cv=10, scoring='accuracy')
    acc_score.append(scores.mean())
print(acc_score)    
```

```python
C_values=list(range(1,26))

plt.plot(C_values,acc_score)
plt.xticks(np.arange(0,27,2))
plt.xlabel('Value of C for SVC')
plt.ylabel('Cross-Validated Accuracy')
plt.grid()
```

```python
C_range=list(np.arange(0.1,4,0.1))
acc_score=[]
for c in C_range:
    svc = SVC(kernel='linear', C=c)
    scores = cross_val_score(svc, X, y, cv=10, scoring='accuracy')
    acc_score.append(scores.mean())
print(acc_score) 
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

C_values=list(np.arange(0.1,4,0.1))
plt.plot(C_values,acc_score)
plt.xticks(np.arange(0.0,4,0.3))
plt.xlabel('Value of C for SVC ')
plt.ylabel('Cross-Validated Accuracy')
plt.grid()
```

optimized value of C=0.6

### Optimize parameter Gamma

```python
gamma_range=[0.0001,0.001,0.01,0.1,1,10,100]
acc_score=[]
for g in gamma_range:
    svc = SVC(kernel='rbf', gamma=g)
    scores = cross_val_score(svc, X, y, cv=10, scoring='accuracy')
    acc_score.append(scores.mean())
print(acc_score)    
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

gamma_range=[0.0001,0.001,0.01,0.1,1,10,100]

sns.lineplot(x=gamma_range,y=acc_score)
plt.xlabel('Value of gamma for SVC ')
plt.xticks(np.arange(0.0001,100,5))
plt.ylabel('Cross-Validated Accuracy')
plt.grid()
```

Gamma optimized value be between 0 and 1.5
